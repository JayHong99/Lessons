### 라인 플롯

```python
plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.show()
```

```python
plt.title("앞의 Plot에 특징들을 추가하기")
plt.plot(
        [10, 20, 30, 40], [1, 4, 9, 16], 
         color = 'r',
         ls = '--',
         marker = 'o',
         markersize = 9,
         markeredgecolor = 'b',
         markerfacecolor = 'b'
        )
plt.xlabel('X축')
plt.ylabel('Y축')
plt.show()
```

```python
plt.title("앞의 Plot에 축의 범위 바꾸기")
plt.plot(
        [10, 20, 30, 40], [1, 4, 9, 16], 
         color = 'r',
         ls = '--',
         marker = 'o',
         markersize = 9,
         markeredgecolor = 'b',
         markerfacecolor = 'b'
        )
plt.xlabel('X축 (0~45)')
plt.ylabel('Y축 (0~20)')

plt.xlim(0,45)
plt.ylim(0,20)

plt.show()
```

```python
plt.title("앞의 Plot에 축의 이름 바꾸고, 그리드 추가하기")
plt.plot(
        [10, 20, 30, 40], [1, 4, 9, 16], 
         color = 'r',
         ls = '--',
         marker = 'o',
         markersize = 9,
         markeredgecolor = 'b',
         markerfacecolor = 'b'
        )
plt.xlabel('X축 (0~45)')
plt.ylabel('Y축 (0~20)')

plt.xlim(0,45)
plt.ylim(0,20)

plt.xticks([10,20,30,40])
plt.yticks([1,4,9,16])
plt.grid(True)

plt.show()
```

```python
plt.title("앞의 Plot에 여러개의 선 그리기")
plt.plot(
        [10, 20, 30, 40], [1, 4, 9, 16], 
         color = 'r',
         ls = '--',
         marker = 'o',
         markersize = 9,
         markeredgecolor = 'b',
         markerfacecolor = 'b'
        )
plt.plot(
        [10, 20, 30, 40], [16, 9, 4, 1], 
         color = 'b',
         ls = ':',
         marker = 'x',
         markersize = 9,
         markeredgecolor = 'r',
         markerfacecolor = 'r'
        )
plt.plot([25,25,25,25], [1,4,9,16], color = 'g', ls = '-')


plt.xlabel('X축 (0~45)')
plt.ylabel('Y축 (0~20)')

plt.xlim(0,45)
plt.ylim(0,20)

plt.xticks([10,20,30,40])
plt.yticks([1,4,9,16])
plt.grid(True)

plt.show()
```

```python
plt.title("앞의 Plot에 텍스트 적기")
plt.plot(
        [10, 20, 30, 40], [1, 4, 9, 16], 
         color = 'r',
         ls = '--',
         marker = 'o',
         markersize = 9,
         markeredgecolor = 'b',
         markerfacecolor = 'b'
        )
plt.plot(
        [10, 20, 30, 40], [16, 9, 4, 1], 
         color = 'b',
         ls = ':',
         marker = 'x',
         markersize = 9,
         markeredgecolor = 'r',
         markerfacecolor = 'r'
        )
plt.plot([25,25,25,25], [1,4,9,16], color = 'g', ls = '-')


plt.xlabel('X축 (0~45)')
plt.ylabel('Y축 (0~20)')

plt.xlim(0,45)
plt.ylim(0,20)

plt.xticks([10,20,30,40])
plt.yticks([1,4,9,16])
plt.grid(True)

plt.text(25 , 6.5, 'o', size = 20,  horizontalalignment='center')
plt.text(27 , 6.5, '겹치는 지점', size = 20, )

plt.show()
```

## 바차트

```python
y = [2, 3, 1]
x = [0, 1, 2]
xlabel = ['가', '나', '다']

plt.title("Bar Chart")
plt.bar(x, y)

plt.xticks(x, xlabel)
plt.yticks(sorted(y))

plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
```

```python
people = ['몽룡', '춘향', '방자', '향단']
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

plt.title("Barh Chart")
plt.barh(y_pos, performance, alpha=0.4) #투명도 조정
plt.yticks(y_pos, people)
plt.xlabel('x 라벨')
plt.show()
```

## 파이 차트

```python
labels = ['개구리', '돼지', '개', '통나무']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes,  # 개별 값들의 크기
        labels=labels,# 개별 값들의 이름 
        colors=colors, # 개별 값들의 색상
        explode=explode, # 0보다 크면, 값이 떨어져서 보임 (돼지 참고)
        autopct='%1.1f%%', # 소수점 몇재짜리까지 보이게 할 것인가?
        shadow=True, # 그림자 추가 여부
        startangle=90) # 시작하는 각도 변경
plt.show()
```

## 히스토그램

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.title("Histogram")
plt.hist(x, bins=10, edgecolor='w')
plt.show()
```

# 스캐터 플롯

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title("Scatter Plot")
plt.scatter(X, Y)
plt.show()
```

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2
plt.title("Bubble Chart")
plt.scatter(x, y1,
            c=y2, # 색상
            s=y3) # 사이즈
plt.show()
```
